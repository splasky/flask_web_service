# Weather

```python
# import xml datasets
# Wx(天氣現象)、MaxT(最高溫度)、MinT(最低溫度)
# https://data.gov.tw/dataset/9219

import urllib
import xml.etree.ElementTree as ET
    
weather_file=urllib.request.urlopen('http://opendata.cwb.gov.tw/govdownload?dataid=F-C0032-005&authorizationkey=rdec-key-123-45678-011121314')
weather_data=weather_file.read()
weather_file.close()
```

```python
root=ET.fromstring(weather_data)
for child in root:
    print(str(child.tag))
```

```python
xml_namespace={'default':'urn:cwb:gov:tw:cwbcommon:0.1'}
dataset=root.find('default:dataset',xml_namespace)
```

```python
import json

class Time_element:
    
    def __init__(self,start_time:str,end_time:str,parameterName:str,parameterValue:str):
        self.start_time=start_time
        self.end_time=end_time
        self.parameterName=parameterName
        self.parameterValue=parameterValue
        
class Weather_element:
    
    def __init__(self,element_name:str):
        self.element_name=element_name
        self.time_element=[]
    def append(self,time_element:Time_element):
        self.time_element.append(time_element)
    
    def __str__(self):
        return self.element_name
        
        
class Weather:
    
    def __init__(self,city:str,Wx:Weather_element,MinT:Weather_element,MaxT:Weather_element):
        self.city=city
        self.Wx=Wx
        self.MinT=MinT
        self.MaxT=MaxT
    
    def __str__(self):
        return str(self.city)
    

for location in dataset.findall('default:location',xml_namespace):
    city=location.find('default:locationName',xml_namespace).text
    weth_eles=[]
    
    for w_ele in location.findall('default:weatherElement',xml_namespace):
        ele_name=w_ele.find('default:elementName',xml_namespace).text
        weth_ele = Weather_element(ele_name)
        if ele_name =='Wx':
            parameterValueRe='default:parameterValue'
        else:
            parameterValueRe='default:parameterUnit'
            
        
        for time in w_ele.findall('default:time',xml_namespace):
                start_time=time.find('default:startTime',xml_namespace).text
                end_time=time.find('default:endTime',xml_namespace).text
                parameter=time.find('default:parameter',xml_namespace)
                parameterName=parameter.find('default:parameterName',xml_namespace).text
                parameterValue=parameter.find(parameterValueRe,xml_namespace).text
                print('{} {},{},{},{},{}'.format(ele_name,city,start_time,end_time,parameterName,parameterValue))
          
```

# 影響空氣品質的因素

```python
# 各空品區污染源管制後排放量一覽表
local_pollution=pd.read_json('http://opendata.epa.gov.tw/ws/Data/ATM00734/?$skip=0&$top=1000&format=json')
local_pollution.head()
```

```python
local_pollution[['area']].drop_duplicates()
```

```python
east=local_pollution[local_pollution['area']=='中部空品區']
east=east[["PM10_EMI","PM2.5_EMI","comp_kind"]]
import numpy as np
east=east.replace("***",np.nan).replace("",np.nan)
east = east[pd.notnull(east['comp_kind'])]
```

```python
east= east[pd.notnull(east['PM2.5_EMI'])]
east= east[pd.notnull(east['PM10_EMI'])]
east["PM10_EMI"]=pd.to_numeric(east["PM10_EMI"])
east["PM2.5_EMI"]=pd.to_numeric(east["PM2.5_EMI"])
```

```python
east=east.groupby('comp_kind').sum()
```

```python
large = east.nlargest(10,"PM2.5_EMI")
```

```python
import json
```

```python
json.dumps({
        "source":list(large.index.get_values()),
        "PM2.5":list(large["PM2.5_EMI"]),
        "PM10":list(large["PM10_EMI"])
    })
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
plt.plot(large.index.get_values(),large["PM2.5_EMI"],large.index.get_values(),large["PM10_EMI"])
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = large.index.get_values()
sizes = large["PM2.5_EMI"]

fig1, ax1 = plt.subplots()
ax1.pie(sizes,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
